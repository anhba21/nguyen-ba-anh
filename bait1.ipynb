```python
#bait1
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt

# To th mc lu nh nu cha c
os.makedirs('output_images', exist_ok=True)

# c nh gc
img = cv2.imread('bird.jpg')

# Tch cc knh mu
b, g, r = cv2.split(img)
zero = np.zeros_like(b)

# nh ch c knh 
red_img = cv2.merge([zero, zero, r])
cv2.imwrite('output_images/red_image.png', red_img)

# nh ch c knh xanh l
green_img = cv2.merge([zero, g, zero])
cv2.imwrite('output_images/green_image.png', green_img)

# nh ch c knh xanh dng
blue_img = cv2.merge([b, zero, zero])
cv2.imwrite('output_images/blue_image.png', blue_img)

# Hin th nh
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(cv2.cvtColor(red_img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Red Channel')
axs[1].imshow(cv2.cvtColor(green_img, cv2.COLOR_BGR2RGB))
axs[1].set_title('Green Channel')
axs[2].imshow(cv2.cvtColor(blue_img, cv2.COLOR_BGR2RGB))
axs[2].set_title('Blue Channel')
for ax in axs:
    ax.axis('off')
plt.show()
```

```python
#bai2
import cv2
import os
import matplotlib.pyplot as plt

os.makedirs('output_swapped_colors', exist_ok=True)
img = cv2.imread('bird.jpg')

# 1. Hon i R v G
rg_img = img.copy()
rg_img[:, :, [1, 2]] = rg_img[:, :, [2, 1]]
cv2.imwrite('output_swapped_colors/swap_rg.jpg', rg_img)

# 2. Hon i G v B
gb_img = img.copy()
gb_img[:, :, [0, 1]] = gb_img[:, :, [1, 0]]
cv2.imwrite('output_swapped_colors/swap_gb.jpg', gb_img)

# 3. Hon i B v R
br_img = img.copy()
br_img[:, :, [0, 2]] = br_img[:, :, [2, 0]]
cv2.imwrite('output_swapped_colors/swap_br.jpg', br_img)

# Hin th nh
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(cv2.cvtColor(rg_img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Swap R-G')
axs[1].imshow(cv2.cvtColor(gb_img, cv2.COLOR_BGR2RGB))
axs[1].set_title('Swap G-B')
axs[2].imshow(cv2.cvtColor(br_img, cv2.COLOR_BGR2RGB))
axs[2].set_title('Swap B-R')
for ax in axs:
    ax.axis('off')
plt.show()
```

```python
#bai3
import cv2
import numpy as np
import os

# To th mc  lu nh kt qu
os.makedirs('output_modified_hsv', exist_ok=True)
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

os.makedirs('output_hsv_images', exist_ok=True)
img = cv2.imread('bird.jpg')
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_img)

hue_img = cv2.cvtColor(cv2.merge([h, 255 * np.ones_like(s), 255 * np.ones_like(v)]), cv2.COLOR_HSV2BGR)
cv2.imwrite('output_hsv_images/hue_channel.jpg', hue_img)
sat_img = cv2.cvtColor(cv2.merge([90 * np.ones_like(h), s, 255 * np.ones_like(v)]), cv2.COLOR_HSV2BGR)
cv2.imwrite('output_hsv_images/saturation_channel.jpg', sat_img)
val_img = cv2.cvtColor(cv2.merge([90 * np.ones_like(h), 255 * np.ones_like(s), v]), cv2.COLOR_HSV2BGR)
cv2.imwrite('output_hsv_images/value_channel.jpg', val_img)

# Hin th nh
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(cv2.cvtColor(hue_img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Hue Channel')
axs[1].imshow(cv2.cvtColor(sat_img, cv2.COLOR_BGR2RGB))
axs[1].set_title('Saturation Channel')
axs[2].imshow(cv2.cvtColor(val_img, cv2.COLOR_BGR2RGB))
axs[2].set_title('Value Channel')
for ax in axs:
    ax.axis('off')
plt.show()
# c nh gc
img = cv2.imread('bird.jpg')

# Chuyn nh sang h mu HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Tch cc knh H, S, V
h, s, v = cv2.split(hsv_img)

# p dng bin i theo  bi
# H (Hue) trong OpenCV nm trong khong 0-179
# => cn m bo sau khi chia 3 vn nm trong khong 0-179
h_new = np.clip((h / 3), 0, 179).astype(np.uint8)
v_new = np.clip((v * 0.75), 0, 255).astype(np.uint8)

# Gp li cc knh  chnh sa vi S gi nguyn
hsv_modified = cv2.merge([h_new, s, v_new])

# Chuyn li sang nh RGB (hoc BGR)  lu
result_img = cv2.cvtColor(hsv_modified, cv2.COLOR_HSV2BGR)

# Lu nh kt qu
cv2.imwrite('output_modified_hsv/modified_hsv_image.jpg', result_img)

print(" lu nh vi H_new = 1/3 * H v V_new = 3/4 * V.")
```

```python
#bai4
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

os.makedirs('output_modified_hsv', exist_ok=True)
img = cv2.imread('bird.jpg')
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_img)

h_new = np.clip((h / 3), 0, 179).astype(np.uint8)
v_new = np.clip((v * 0.75), 0, 255).astype(np.uint8)
hsv_modified = cv2.merge([h_new, s, v_new])
result_img = cv2.cvtColor(hsv_modified, cv2.COLOR_HSV2BGR)
cv2.imwrite('output_modified_hsv/modified_hsv_image.jpg', result_img)

# Hin th nh
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
plt.title('Modified HSV Image')
plt.axis('off')
plt.show()
```

```python
#bai5
import cv2
import os
import matplotlib.pyplot as plt

input_folder = r'C:\VScode\baitap t1'
output_folder = 'Exercise_mean_filtered'
os.makedirs(output_folder, exist_ok=True)

kernel_size = (5, 5)
example_img = None

for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
        image_path = os.path.join(input_folder, filename)
        img = cv2.imread(image_path)
        mean_filtered = cv2.blur(img, kernel_size)
        output_path = os.path.join(output_folder, f'mean_{filename}')
        cv2.imwrite(output_path, mean_filtered)
        if example_img is None:
            example_img = mean_filtered

print(" p dng mean filter v lu nh vo th mc:", output_folder)

# Hin th mt nh v d
if example_img is not None:
    plt.figure(figsize=(5, 5))
    plt.imshow(cv2.cvtColor(example_img, cv2.COLOR_BGR2RGB))
    plt.title('Mean Filter Example')
    plt.axis('off')
    plt.show()
```
